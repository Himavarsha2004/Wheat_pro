```python
import pandas as pd 
df=pd.read_csv("/kaggle/input/genotype-data-csv/Genotypic_Data.csv") 
df
```

```python
def replace_none_with_first_char(row, column_name):
    if pd.isna(row[column_name]):
        return row['alleles'][2]
    else:
        return row[column_name]

# List of columns to process
columns_to_process = df.columns

# Apply the function to each specified column in the DataFrame
for column in columns_to_process:
    df[column] = df.apply(replace_none_with_first_char, axis=1, column_name=column)

# Display the updated DataFrame
rep={'AA': 1, 'AC': 2, 'AG': 3,'AT': 4, 'CC': 5, 'CG': 6,'TC': 7, 'TG': 8, 'TT': 9,'GG':10, 'A':11, 'C':12, 'G':13, 'T':14}
df=df.replace(rep)
df
```

```python
columns_to_drop = ['SNPs','alleles','Chrom','Pos','strand','assembly','center','protLSID','assayLSID','panel','QCcode']
df1= df.drop(columns=columns_to_drop)
```

```python
df_transposed = df1.T
df_transposed.reset_index(inplace=True)
df_transposed.rename(columns={'index': 'Genotype'}, inplace=True)

df_transposed
```
